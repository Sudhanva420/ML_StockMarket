```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
import warnings
import numpy as np
import os 
os.environ['KMP_DUPLICATE_LIB_OK']='True'
warnings.filterwarnings('ignore')
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
start_date = "2020-01-01"
end_date = "2020-08-01"
infy_data=yf.download("INFY",start=start_date,end=end_date)
infy_data
```

```python
infy_data=infy_data.dropna()
infy_data
```

```python
'''

Within each iteration of the loop, a new column named 'Close_lag' is created in the infy_data DataFrame. 
This column is populated with values from the 'Close' column, but shifted by the current lag value. 
For example, when lag is 1, 'Close_lag' will contain the 'Close' values from the previous day; when lag is 2, it will contain values 
from two days ago, and so on.
'''

for lag in range(1, 8):  # Using the past 7 days as features
    infy_data['Close_lag'] = infy_data['Close'].shift(lag)

infy_data.dropna(inplace=True)
infy_data
```

```python
infy_data=infy_data.drop('Close',axis='columns')
infy_data
```
