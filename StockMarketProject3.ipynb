```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
import warnings
import numpy as np
import os 
os.environ['KMP_DUPLICATE_LIB_OK']='True'
warnings.filterwarnings('ignore')
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
start_date = "2020-01-01"
end_date = "2020-08-01"
infy_data=yf.download("INFY",start=start_date,end=end_date)
infy_data
```

```python
infy_data=infy_data.dropna()
infy_data=infy_data.drop(index=infy_data.index[56])
infy_data=infy_data.drop(index=infy_data.index[96])
infy_data
```

```python
start_date = "2020-08-02"
end_date = "2021-03-02"
yx=yf.download("INFY",start=start_date,end=end_date)
y=yx['Close']
y
#y.columns
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

xtrain1=scaler.fit_transform(infy_data)
xtrain1
```

```python
y=pd.DataFrame(y)
ytrain1=scaler.fit_transform(y)
ytrain1
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(xtrain1,ytrain1,test_size=0.3)
y_test.shape
```

```python
ytrain1
```

```python
y_train=pd.DataFrame(y_train)
y_train

# Rename the column
y_train1 = y_train.rename(columns={0: 'Close'})

y_train1.columns
```

```python
#Random Forest

import tensorflow_decision_forests as tfdf
x_train=pd.DataFrame(x_train,columns=infy_data.columns)
x_train['Close2'] = y_train1['Close']
x_train
```

```python
label='Close2'
train_ds=tfdf.keras.pd_dataframe_to_tf_dataset(x_train,label='Close2',task=tfdf.keras.Task.REGRESSION)
```

```python
rf=tfdf.keras.RandomForestModel(task=tfdf.keras.Task.REGRESSION)
rf.compile(metrics=["mse"])
```

```python
rf.fit(x=train_ds)
```

```python
y_test=pd.DataFrame(y_test)
y_test

# Rename the column
y_test1 = y_test.rename(columns={0: 'Close'})

y_test1.columns
```

```python
x_test=pd.DataFrame(x_test,columns=infy_data.columns)
x_test['Close2'] = y_test1['Close']
x_test
```

```python
test_ds=tfdf.keras.pd_dataframe_to_tf_dataset(x_test,label='Close2',task=tfdf.keras.Task.REGRESSION)
```

```python
preds=rf.predict(x=test_ds)
```

```python
actual_data=x_test['Close2']
actual_data

```

```python
import matplotlib.pyplot as plt
plt.plot(actual_data)
plt.plot(preds)
plt.legend(('Data','Prediction'))
```
