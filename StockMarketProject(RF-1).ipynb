```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
import warnings
import numpy as np
import os 
os.environ['KMP_DUPLICATE_LIB_OK']='True'
warnings.filterwarnings('ignore')
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
start_date = "2020-01-01"
end_date = "2020-08-01"
infy_data=yf.download("INFY",start=start_date,end=end_date)
infy_data
```

```python
infy_data=infy_data.dropna()
infy_data=infy_data.drop(index=infy_data.index[56])
infy_data=infy_data.drop(index=infy_data.index[96])
infy_data
```

```python
start_date = "2020-08-02"
end_date = "2021-03-02"
yx=yf.download("INFY",start=start_date,end=end_date)
y=yx['Close']
y
#y.columns
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

xtrain1=scaler.fit_transform(infy_data)
xtrain1
```

```python
y=pd.DataFrame(y)
ytrain1=scaler.fit_transform(y)
ytrain1
```

```python
y_train=pd.DataFrame(ytrain1)
y_train

# Rename the column
y_train = y_train.rename(columns={0: 'Close'})

y_train.columns
```

```python
#Random Forest

import tensorflow_decision_forests as tfdf
x_train=pd.DataFrame(xtrain1,columns=infy_data.columns)
x_train['Close2'] = y_train['Close']
x_train
```

```python
label='Close2'
train_ds=tfdf.keras.pd_dataframe_to_tf_dataset(x_train,label='Close2',task=tfdf.keras.Task.REGRESSION)
```

```python
rf=tfdf.keras.RandomForestModel(task=tfdf.keras.Task.REGRESSION)
rf.compile(metrics=["mse"])
```

```python
rf.fit(x=train_ds)
```

```python
start_date = "2021-01-01"
end_date = "2021-08-01"
infy_test=yf.download("INFY",start=start_date,end=end_date)
infy_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

xtest1=scaler.fit_transform(infy_test)
xtest1
```

```python
x_test=pd.DataFrame(xtest1,columns=infy_test.columns)
x_test
```

```python
test_ds = tfdf.keras.pd_dataframe_to_tf_dataset(x_test, task=tfdf.keras.Task.REGRESSION)
```

```python
predictions = rf.predict(x=test_ds)
predictions_df = pd.DataFrame(predictions)
predictions.shape
```

```python
start_date = "2021-08-02"
end_date = "2022-03-02"
infy_actual=yf.download("INFY",start=start_date,end=end_date)
infy_actual
```

```python
infy_close=infy_actual['Close']
infy_close=infy_close.drop(index=infy_close.index[56])
infy_close=infy_close.drop(index=infy_close.index[96])
infy_c=pd.DataFrame(infy_close)
infy_c
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

xactual1=scaler.fit_transform(infy_c)
xactual1=pd.DataFrame(xactual1)
xactual1
```

```python
import matplotlib.pyplot as plt
plt.plot(xactual1)
plt.plot(predictions)
plt.legend(('Actual Data','Predcitions'),fontsize=15)
```

```python
predictions
predictions_df = pd.DataFrame(predictions, columns=['Predicted_Close'])
predictions_df.index = infy_test.index 
predictions_df
```

```python
actual_df3 = pd.DataFrame(xactual1)
actual_df3.index = infy_test.index
actual_df3
```

```python
predictions_df
```

```python
plt.figure(figsize=(14, 7))
plt.plot(actual_df3, label='Actual Data', color='blue')
plt.plot(predictions_df['Predicted_Close'], label='Predictions', linestyle='--', color='red')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Actual vs Predicted Close Price')
plt.legend(fontsize=15)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mse=mean_squared_error(actual_df3,predictions_df['Predicted_Close'])
mae=mean_absolute_error(actual_df3,predictions_df['Predicted_Close'])
r2_score=r2_score(actual_df3,predictions_df['Predicted_Close'])
```
